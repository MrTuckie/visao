###2- Determine se o ponto "p" pertence ou não à reta "r":

a) r = (3, 5, 2), p = (1, -1, 1)

b) r = (3, 5, 2), p = (1, 0, 1)

c) r = (3, 5, 2), p = (1, 1, 1)

d) r = (2, 1, -6), p = (-2, 0, 1)

e) r = (2, 1, -6), p = (-2, -4, 1)

f) r = (-4, 7, 0), p = (5, -3, 1)

g) r = (-4, 7, 0), p = (14, 28, 3)

h) r = (0, 0, 1), p = (0, 0, 1)

i) r = (0, 0, 1), p = (4, 0, 0)

Siga o exemplo de código abaixo:

```
#reta
r = np.array([3, 3, 1])
#ponto
p = np.array([1, 1, 2])
#fazer o produto interno e verificar se dá zero
d = np.dot(r,p)
print(d)
if d == 0:
    print("O ponto pertence à reta.")
else:
    print("O ponto não pertence à reta.")
```

###3-Determine a reta que passa pelos pontos:

a) (2, 3, 1) e (5, 6, 4)

b) (-3, 1, -2) e (3, 2, -1)

c) (8, -3, -1) e (-7, 9, 0)

Siga o exemplo de código abaixo:


```
#pontos
p1 = np.array ([10,-3, 1])
p2 = np.array ([-3, 5, 1])

#calcula a reta

r = np.cross(p1,p2)
print("A reta é ", r)
```
